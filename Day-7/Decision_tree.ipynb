```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=['sepal length','sepal width','petal length','petal width'])

df['target'] = iris.target

df['target'] = df['target'].apply(lambda x: iris.target_names[x])

df.head()
x=df.drop('target',axis=1)
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
label = LabelEncoder()
temp1=label.fit_transform(y_train)
temp2=label.transform(y_test)
y_train = temp1
y_test = temp2
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision = DecisionTreeClassifier()
```

```python
decision.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(decision,filled=True)
```

```python
y_pred = decision.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,precision_score,confusion_matrix
```

```python
accutracy = accuracy_score(y_test,y_pred)
print(accutracy)
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

# Handling Categorical Features

### There are three methods to handle categorical values
#### 1. Dropping The categorical columns
#### 2. Ordinal Encoding
#### 3. One-Hot Encoding

```python
# 1. Dropping The categorical columnsÂ¶

column = [col for col in data.columns if type(col)=='object']
column
dropped_data = data.select_dtypes(exclude=['object'])
dropped_data.head()

# second method

column = [col for col in data.columns if data[col].dtype == 'object']
data2=data.drop(columns=column)
data2
```

```python
# 2. Ordinal Encoding

from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
cat = (data.dtypes == 'object')
list_cat = list(cat[cat].index)
# list_cat

data1 = data.copy()
data2 = data.copy()

## Apply Ordinal Encoding over categorical features

data1[list_cat] = ord_enc.fit_transform(data1[list_cat])
data2[list_cat] = ord_enc.transform(data2[list_cat])
```

```python
# 3. One-Hot Encoding

from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

data3 = data.copy()
data4 = data.copy()

data3 = pd.DataFrame(one_hot_encoder.fit_transform(data3[list_cat]))
data4 = pd.DataFrame(one_hot_encoder.transform(data4[list_cat]))
```

## Data Imputation

```python
# 1. groupby Method (fill the value with similer item of value in feature)
car_data['Engine Fuel Type'] = car_data.groupby('Model')['Engine Fuel Type'].transform(lambda x:x.fillna(x.mode().iloc[0] if not x.mode().empty else None))
car_data['Number of Doors'] = car_data.groupby('Model')['Number of Doors'].transform(lambda x:x.fillna(x.mean()))
car_data['Engine Cylinders'] = car_data.groupby('Model')['Engine Cylinders'].transform(lambda x:x.fillna(x.mean()))
car_data['Engine HP'] = car_data.groupby('Model')['Engine HP'].transform(lambda x:x.fillna(x.mean()))

car_data

# 2. Rule-Based Imputation
```
